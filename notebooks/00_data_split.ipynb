```python
import pandas as pd
import numpy as np

np.random.seed(42)
```

```python
pod_meta = pd.read_excel("../data/raw/podlesny2022_metadata.xlsx")
```

```python
print("Code is running")
```

```python
all_cases = pod_meta['Case_Name'].unique()
n_test = int(len(all_cases) * 0.20)

test_cases = np.random.choice(all_cases, size=n_test, replace=False)
train_cases = [c for c in all_cases if c not in test_cases]

print(f"Total cases:  {len(all_cases)}")
print(f"Train cases:  {len(train_cases)} (80%)")
print(f"Test cases:   {len(test_cases)} (20%)")
```

```python
pod_meta['split'] = pod_meta['Case_Name'].apply(
    lambda c: 'test' if c in test_cases else 'train'
)
```

```python
pd.Series(train_cases).to_csv(
    "../data/processed/train_cases.csv", index=False
)
pd.Series(test_cases).to_csv(
    "../data/processed/test_cases.csv", index=False
)
pod_meta.to_csv(
    "../data/processed/podlesny_meta_split.csv", index=False
)

print("\n✓ Split locked and saved")
print("RULE: Smillie mmc6 = external benchmark ONLY, never train")
```
