```python
import pandas as pd

hmp2_raw = pd.read_csv(
    "../data/raw/HMP2_taxonomic_profiles.tsv.gz",
    sep="\t",
    index_col=0,
    compression="gzip"
)
```

```python
hmp2 = hmp2_raw.T

matched_ids = hmp2.index.intersection(meta_wgs.index)

print("Matched samples:", len(matched_ids))
```

```python
cd_samples = meta_wgs[meta_wgs['diagnosis'] == 'CD'].index
uc_samples = meta_wgs[meta_wgs['diagnosis'] == 'UC'].index
healthy_samples = meta_wgs[meta_wgs['diagnosis'] == 'nonIBD'].index

hmp2_cd = hmp2.loc[cd_samples]
hmp2_uc = hmp2.loc[uc_samples]
hmp2_healthy = hmp2.loc[healthy_samples]

print("CD:", hmp2_cd.shape)
print("UC:", hmp2_uc.shape)
print("Healthy:", hmp2_healthy.shape)
```

```python
def filter_rare(df, threshold=0.05):
    keep = (df > 0).mean(axis=0) > threshold
    return df.loc[:, keep]

hmp2_cd = filter_rare(hmp2_cd)
hmp2_uc = filter_rare(hmp2_uc)
hmp2_healthy = filter_rare(hmp2_healthy)

print(hmp2_cd.shape, hmp2_uc.shape, hmp2_healthy.shape)
```

```python
import numpy as np

hmp2_cd = np.log1p(hmp2_cd)
hmp2_uc = np.log1p(hmp2_uc)
hmp2_healthy = np.log1p(hmp2_healthy)
```

```python
common_species = (
    set(hmp2_cd.columns)
    & set(hmp2_uc.columns)
    & set(hmp2_healthy.columns)
)

common_species = list(common_species)

hmp2_cd = hmp2_cd[common_species]
hmp2_uc = hmp2_uc[common_species]
hmp2_healthy = hmp2_healthy[common_species]

print(hmp2_cd.shape, hmp2_uc.shape, hmp2_healthy.shape)
```

```python
hmp2_matched = hmp2.loc[matched_ids]
meta_matched = meta_wgs.loc[matched_ids]
```

```python
print("\nSplitting by diagnosis...")

hmp2_matched = hmp2.loc[matched_ids]
meta_matched = meta_wgs.loc[matched_ids]

cd_idx = meta_matched[meta_matched['diagnosis'] == 'CD'].index
uc_idx = meta_matched[meta_matched['diagnosis'] == 'UC'].index
healthy_idx = meta_matched[meta_matched['diagnosis'] == 'nonIBD'].index

hmp2_cd = hmp2_matched.loc[cd_idx]
hmp2_uc = hmp2_matched.loc[uc_idx]
hmp2_healthy = hmp2_matched.loc[healthy_idx]

print(f"CD samples: {len(hmp2_cd)}")
print(f"UC samples: {len(hmp2_uc)}")
print(f"Healthy samples: {len(hmp2_healthy)}")
```

```python
import pandas as pd
import numpy as np

def preprocess_condition(condition_name, apply_prevalence=0.20):

    print(f"\n{'='*45}")
    print(f"PROCESSING: {condition_name}")
    print('='*45)

    df = pd.read_csv(
        f"../data/processed/{condition_name}_raw.csv",
        index_col=0
    )
    print(f"Loaded: {df.shape}")

    if df.columns.str.contains('s__').any():
        species_cols = df.columns[
            df.columns.str.contains('s__') &
            ~df.columns.str.contains('t__')
        ]
        df = df[species_cols]
        df.columns = df.columns.str.split('s__').str[-1]

    print(f"Species level: {df.shape[1]} species")

    if df.values.max() > 1.0:
        df = df / 100.0

    prev = (df > 0).mean(axis=0)
    df = df.loc[:, prev >= apply_prevalence]
    print(f"After prevalence filter: {df.shape[1]} species")

    df = df.loc[:, df.mean(axis=0) >= 0.0001]
    print(f"After abundance filter: {df.shape[1]} species")

    sums = df.sum(axis=1)
    Q1, Q3 = sums.quantile([0.25, 0.75])
    IQR = Q3 - Q1

    outliers = sums[
        (sums < Q1 - 3*IQR) | (sums > Q3 + 3*IQR)
    ]

    df = df.drop(index=outliers.index)
    print(f"After outlier removal: {df.shape[0]} samples")

    df_norm = df.div(df.sum(axis=1), axis=0)

    df_pseudo = df_norm + 1e-6
    log_df = np.log(df_pseudo)
    df_clr = log_df.sub(log_df.mean(axis=1), axis=0)

    df_norm.to_csv(f"../data/processed/{condition_name}_normalized.csv")
    df_clr.to_csv(f"../data/processed/{condition_name}_CLR.csv")

    print("Saved ")

    return df_norm, df_clr
```

```python
cd_norm, cd_clr = preprocess_condition("CD")
uc_norm, uc_clr = preprocess_condition("UC")
healthy_norm, healthy_clr = preprocess_condition("healthy")
```

```python
common_species = set(cd_norm.columns) & set(uc_norm.columns) & set(healthy_norm.columns)

print("Common species:", len(common_species))

cd_norm = cd_norm[list(common_species)]
uc_norm = uc_norm[list(common_species)]
healthy_norm = healthy_norm[list(common_species)]

cd_norm.to_csv("../data/processed/CD_aligned.csv")
uc_norm.to_csv("../data/processed/UC_aligned.csv")
healthy_norm.to_csv("../data/processed/healthy_aligned.csv")
```
