```python
import pandas as pd

def smart_load(path, nrows=3):
    ext = path.split('.')[-1].lower()
    
    if ext == 'gz':
        df = pd.read_csv(path, sep='\t', compression='gzip', nrows=nrows)
    elif ext == 'tsv':
        df = pd.read_csv(path, sep='\t', nrows=nrows)
    elif ext == 'csv':
        df = pd.read_csv(path, nrows=nrows)
    elif ext == 'xlsx':
        df = pd.read_excel(path, nrows=nrows)
    else:
        print("Unknown file type:", path)
        return None
    
    print("Loaded:", path)
    print(df.head())
    print("-" * 40)
    return df
```
