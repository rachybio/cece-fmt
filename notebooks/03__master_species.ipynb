```python
import pandas as pd
import numpy as np

print("Building master species sets...")

cd = pd.read_csv("../data/processed/CD_normalized.csv", index_col=0)
uc = pd.read_csv("../data/processed/UC_normalized.csv", index_col=0)
healthy = pd.read_csv("../data/processed/healthy_normalized.csv", index_col=0)

pod = pd.read_excel("../data/raw/podlesny2022_abundance.xlsx")
pod_species = set(
    pod['species']
    .astype(str)
    .str.extract(r's__([A-Za-z_]+)', expand=False)
    .fillna(pod['species'])
    .str.replace(' ', '_')
    .str.strip()
)

smillie_species_name_map = pd.read_excel("../data/raw/smillie_species_name_map.xlsx")
smillie_species = set(
    smillie_species_name_map[smillie_species_name_map['Detected'] == 1]['Genome']
    .astype(str)
    .str.extract(r's__([A-Za-z]+)', expand=False)
    .fillna(smillie_species_name_map['Genome'])
    .str.replace(' ', '_')
    .str.strip()
)

network_species = (
    set(cd.columns) &
    set(uc.columns) &
    set(healthy.columns)
)

ml_species = pod_species & smillie_species

cross_species = set(cd.columns) & pod_species

print(f"Network species: {len(network_species)}")
print(f"ML species: {len(ml_species)}")
print(f"Cross species: {len(cross_species)}")

pd.Series(sorted(network_species)).to_csv(
    "../data/processed/MASTER_network_species.csv",
    index=False, header=False
)

pd.Series(sorted(ml_species)).to_csv(
    "../data/processed/MASTER_ml_species.csv",
    index=False, header=False
)

pd.Series(sorted(cross_species)).to_csv(
    "../data/processed/MASTER_cross_species.csv",
    index=False, header=False
)

print("Master species files updated ")

net_sp = sorted(network_species)

cd[net_sp].to_csv("../data/processed/CD_network.csv")
uc[net_sp].to_csv("../data/processed/UC_network.csv")
healthy[net_sp].to_csv("../data/processed/healthy_network.csv")

print("Master species sets saved")
```
