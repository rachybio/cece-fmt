```python
import pandas as pd

datasets = {
    "CD": "/content/fastspar/sparcc_CD/pvals_CD.csv",
    "UC": "/content/fastspar/sparcc_UC/pvals_UC.csv",
    "Healthy": "/content/fastspar/sparcc_healthy/pvals_healthy.csv"
}

for name, path in datasets.items():
    print(f"\n==== {name} ====")

    df = pd.read_csv(path, sep='\t', index_col=0, comment='#')
    df = df.apply(pd.to_numeric, errors='coerce')

    print("Min, Max:", df.min().min(), df.max().max())
    print("Significant (<0.05):", (df < 0.05).sum().sum())
    print("Non-NaN values:", df.notna().sum().sum())
```

```python
import numpy as np

for name, path in datasets.items():
    df = pd.read_csv(path, sep='\t', index_col=0, comment='#')
    df = df.apply(pd.to_numeric, errors='coerce')

    diff = np.abs(df - df.T)

    print(f"{name} symmetry max diff:", np.nanmax(diff))
```

```python
for name, path in datasets.items():
    df = pd.read_csv(path, sep='\t', index_col=0, comment='#')
    df = df.apply(pd.to_numeric, errors='coerce')

    sig = (df < 0.05)
    print(f"{name}: {sig.sum().sum()} significant edges")
```

```python
for name, path in datasets.items():
    df = pd.read_csv(path, sep='\t', index_col=0, comment='#')
    df = df.apply(pd.to_numeric, errors='coerce')

    total = df.notna().sum().sum()
    sig = (df < 0.05).sum().sum()

    print(f"{name}: {sig/total:.4f} fraction significant")
```
