```python
import pandas as pd
import numpy as np

cd_clr = pd.read_csv("../data/processed/CD_CLR.csv", index_col=0)
meta = pd.read_csv("../data/processed/HMP2_meta_matched.csv", index_col=0)

meta_cd = meta[meta['diagnosis'] == 'CD']

common = cd_clr.index.intersection(meta_cd.index)
df = cd_clr.loc[common].copy()

df['subject_id'] = meta_cd.loc[common, 'Participant ID']
df['timepoint'] = meta_cd.loc[common, 'week_num']

df = df.sort_values(['subject_id', 'timepoint'])

counts = df.groupby('subject_id').size()
valid = counts[counts >= 3].index
df = df[df['subject_id'].isin(valid)]

print("Longitudinal dataset:")
print("  Valid patients:", df['subject_id'].nunique())
print("  Total samples:", len(df))
print("  Week range:", df['timepoint'].min(), "to", df['timepoint'].max())

net_sp = pd.read_csv(
    "../data/processed/MASTER_network_species.csv",
    header=None
)[0].tolist()

species_in_data = [s for s in net_sp if s in df.columns]

df[species_in_data].to_csv(
    "../data/processed/CD_longitudinal_CLR.csv"
)

df.to_csv(
    "../data/processed/CD_longitudinal_with_meta.csv"
)

print("Species retained:", len(species_in_data))
print("\n✓ Longitudinal data ready for Phase 3!")
```
